Quivers&Gradients

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
from osgeo import gdal
import geopandas as gpd
from itertools import repeat
```

```python
arr = np.array([list(repeat(1,5)),
                list(repeat(2,5)),
                list(repeat(3,5)),
                list(repeat(2,5)),
                list(repeat(1,5))],
                dtype=float)
u,v = np.gradient(arr)

x = np.arange(0,5,1)
y = np.arange(0,5,1)

plt.figure(figsize=(15,10))
plt.imshow(arr,  cmap='rainbow')
plt.colorbar()
plt.quiver(x,y,v,u)
plt.show()
```

```python
x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))

u = x/np.sqrt(x**2 + y**2)
v = y/np.sqrt(x**2 + y**2)
u[0][0] = 0
v[0][0] = 0
u[0][1] = v[0][1] = 0
u[0][2] = v[0][2] = 0

plt.quiver(x,y,u,v)
plt.show()
```
